```python
# Импортируем библиотеку pandas

import pandas as pd
```

```python
# Загружаем исходные данные

df = pd.read_csv('XYZ input.csv', delimiter=';')
df
```

```python
# Сохраняем транспонированную таблицу для вычислений

df_t = df.transpose()
df_t
```

```python
# Добавляем столбец с коэффициентом вариации для каждого товара

df['coef_var'] = df_t.iloc[1:, :].std() / df_t.iloc[1:, :].mean()
df
```

```python
# Добавляем столбец с группой товара. 
# Присваиваем позициям ту или иную группу: до 10% - группу X, до 25% - Y, остальное – Z.

df.loc[df.coef_var >= 0.25, 'XYZ-group'] = 'Z'
df.loc[df.coef_var < 0.25, 'XYZ-group'] = 'Y'
df.loc[df.coef_var < 0.1, 'XYZ-group'] = 'X'
df
```

```python
# Добавляем столбец с доходом за все месяцы для каждого товара

df['income'] = df_t.iloc[1:, :].sum()
df
```

```python
# Сортируем данные по убыванию суммарного дохода

df = df.sort_values('income', ascending=False)
df
```

```python
# Считаем суммарный доход

total_income = df['income'].sum()
total_income
```

```python
# Добавляем столбец с долей дохода от каждого товара

df['share'] = df['income']/total_income
df
```

```python
# Добавляем столбец с накопленной долей

df['accum_share'] = df['share'].cumsum()
df
```

```python
# Добавляем столбец с группой товара. 
# Присваиваем позициям ту или иную группу: до 80% - группу А, до 95% - В, остальное – С.

df.loc[df.accum_share <= 1, 'ABC-group'] = 'C'
df.loc[df.accum_share < 0.95, 'ABC-group'] = 'B'
df.loc[df.accum_share < 0.8, 'ABC-group'] = 'A'
df
```

```python
# Записываем результирующую таблицу в excel-файл

df.to_excel('ABC-XYZ output.xlsx') 
```
